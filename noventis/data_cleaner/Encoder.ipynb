```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from category_encoders import TargetEncoder, OrdinalEncoder
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    """
    Calculates Cramér's V statistic for categorical-categorical association.
    """
    # Create a contingency table
    confusion_matrix = pd.crosstab(x, y)

    # Chi-squared test
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    if n == 0:
        return 0

    # Number of rows and columns
    r, k = confusion_matrix.shape

    # Calculate phi^2
    phi2 = chi2 / n

    # Bias correction
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)

    # Avoid division by zero
    if min((kcorr-1), (rcorr-1)) == 0:
        return 0

    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))


def Encoding(data: pd.DataFrame, method: str = 'auto', target_column: str = None, columns_to_encode: list = None, category_mapping: dict = None) -> pd.DataFrame:
    """
    Encodes categorical columns using an automatic or manual method based on statistical properties.

    Args:
        data (pd.DataFrame): The input DataFrame.
        method (str, optional): The encoding method. Defaults to 'auto'.
            'auto': Intelligently selects encoding based on cardinality and correlation with the target.
            'label', 'ohe', 'target', 'ordinal': For manual control.
        target_column (str, optional): The name of the target variable column. Required for 'auto' and 'target' modes.
        columns_to_encode (list, optional): A list of columns to encode. Required for manual modes.
        category_mapping (dict, optional): A mapping dictionary for ordinal encoding.

    Returns:
        pd.DataFrame: The DataFrame with encoded columns.
    """
    if not isinstance(data, pd.DataFrame):
        raise TypeError("Input 'data' must be a pandas DataFrame.")

    df = data.copy()

    ## METHOD AUTO
    if method == 'auto':
        print("--- Running Smart 'auto' mode ---")
        if target_column is None:
            raise ValueError("Parameter 'target_column' is required for 'auto' mode.")
        if target_column not in df.columns:
            raise KeyError(f"Target column '{target_column}' not found in the DataFrame.")


        y = df[target_column]
        # If target is continuous, bin it for Cramér's V calculation
        if pd.api.types.is_numeric_dtype(y) and y.nunique() > 10:
            y_binned = pd.qcut(y, q=4, duplicates='drop', labels=False)
            print(f"Info: Target '{target_column}' is numeric, binning for correlation analysis.")
        else:
            y_binned = y

        cols_to_process = df.select_dtypes(include=['object', 'category']).columns.tolist()
        print(f"Analyzing columns: {cols_to_process}\n")

        for column in cols_to_process:
            unique_count = df[column].nunique()
            print(f"Processing '{column}' (Unique values: {unique_count})")

            # Rule 1: Label Encoding for binary features
            if unique_count == 2:
                print(" -> Decision: Binary feature. Applying Label Encoding.")
                le = LabelEncoder()
                df[f'{column}_label'] = le.fit_transform(df[column])
                df.drop(column, axis=1, inplace=True)
                continue

            # Rule 2: Target Encoding for high cardinality features
            if unique_count > 15:
                print(" -> Decision: High cardinality. Applying Target Encoding.")
                te = TargetEncoder()
                df[f'{column}_target'] = te.fit_transform(df[column], y)
                df.drop(column, axis=1, inplace=True)
                continue

            # Rule 3: OHE vs. Ordinal/Target for low/medium cardinality
            if 3 <= unique_count <= 15:
                correlation = cramers_v(df[column], y_binned)
                print(f" -> Correlation with target (Cramér's V): {correlation:.4f}")

                # Use OHE for features with low correlation
                if correlation < 0.25:
                    print(" -> Decision: Low correlation. Applying One-Hot Encoding.")
                    df = pd.get_dummies(df, columns=[column], prefix=column)
                else:
                # For strong correlation, recommend manual encoding
                    print(f" -> Decision: Strong correlation detected. Skipping automatic encoding.")
                    print(f"    RECOMMENDATION: Manually apply 'ordinal' encoding for '{column}' if categories have a known order,")
                    print(f"    or use 'target' encoding for a robust alternative.")

            print("-" * 20)

        print("\n--- Smart 'auto' mode finished. ---")
        return df

    ## MODE MANUAL
    else:
        # (Manual mode logic remains the same)
        print(f"--- Running manual mode: '{method}' ---")
        if columns_to_encode and not isinstance(columns_to_encode, list):
            raise ValueError("'columns_to_encode' needs to be in list type.")

        if method not in ['label', 'ohe', 'target', 'ordinal']:
            print(f"The method entered: '{method}' is not compatible. Please enter a compatible method: ['label', 'ohe', 'target', 'ordinal'] in manual mode. Returning original data.")

            # return data when the entered method is not compatible
            return data

        # if performing target encoding, the target column must be numeric; convert the target column to numeric
        if method=='target' and not pd.api.types.is_numeric_dtype(df[target_column]):
            print(f"Target column '{target_column}' must be numeric. Converting column {target_column} to numeric.")
            le = LabelEncoder()
            df[f'{target_column}_encoded'] = le.fit_transform(df[target_column])
            df = df.drop([target_column], axis=1)
            target_column = f'{target_column}_encoded'

        # if 'columns_to_encode' is not specified, default to all columns in the dataframe
        if not columns_to_encode:
            columns_to_encode = df.columns

        for column in columns_to_encode:
            # 1. Label Encoding
            if method == 'label':
                if df[column].nunique() > 2:
                    print(f"Column '{column}' has {df[column].nunique()} unique values. Not applying 'label' method to column {column}.")
                    continue
                print(f"Applying Label Encoding on '{column}'...")
                le = LabelEncoder()
                new_col_name = f'{column}_label_encoded'
                df[new_col_name] = le.fit_transform(df[column])

            # 2. One-Hot Encoding
            elif method == 'ohe':
                if df[column].nunique() > 15:
                    print(f"Column '{column}' has {df[column].nunique()} unique values. Not applying 'ohe' method to column {column}.")
                print(f"Applying One-Hot Encoding on '{column}'...")

                df = pd.get_dummies(df, columns=[column], prefix=column, drop_first=False)

            # 3. Target Encoding
            elif method == 'target':
                if column==target_column:
                    continue
                print(f"Applying Target Encoding on '{column}'...")
                if target_column is None:
                    raise ValueError("'target_column' must be specified for target encoding.")
                if target_column not in df.columns:
                    raise KeyError(f"Target column '{target_column}' not found.")

                te = TargetEncoder()
                new_col_name = f'{column}_target_encoded'
                df[new_col_name] = te.fit_transform(df[column], df[target_column])

            # 4. Ordinal Encoding
            elif method == 'ordinal':
                if column not in category_mapping.keys():
                  print(f'Column {column} is not in the mapping provided by the user.')
                  continue
                print(f"Applying Ordinal Encoding on '{column}'...")
                if category_mapping is None or column not in category_mapping:
                    raise ValueError(f"Mapping for column '{column}' not found in 'category_mapping'.")
                if not isinstance(category_mapping[column], dict):
                    raise TypeError(f"Mapping for '{column}' must be a dictionary.")

                unique_vals_in_col = set(df[column].unique())
                mapped_vals = set(category_mapping[column].keys())
                if not unique_vals_in_col.issubset(mapped_vals):
                    unmapped = unique_vals_in_col - mapped_vals
                    raise ValueError(f"Mapping for '{column}' is missing keys for values: {unmapped}")

                # Create the correct mapping format for OrdinalEncoder
                encoder_mapping = [{'col': column, 'mapping': category_mapping[column]}]
                oe = OrdinalEncoder(mapping=encoder_mapping)

                # fit_transform returns the entire df, so we take the relevant column
                df[column] = oe.fit_transform(df)[column]
                df.rename(columns={column: f'{column}_ordinal_encoded'}, inplace=True)

    # return the dataframe with the user-requested encoded columns
    return df


# --- Example Usage ---
if __name__ == "__main__":
    # Create a more complex and realistic dataset
    data = {
        'payment_method': ['Credit Card', 'GoPay', 'Credit Card', 'OVO', 'Bank Transfer', 'GoPay', 'Credit Card', 'OVO', 'OVO', 'Bank Transfer'], # Medium cardinality, strong correlation
        'product_category': ['Electronics', 'Fashion', 'Groceries', 'Books', 'Fashion', 'Electronics', 'Groceries', 'Books', 'Fashion', 'Electronics'], # Medium cardinality, low correlation
        'customer_gender': ['Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male'], # Binary
        'quality': ['good', 'very good', 'bad', 'average', 'bad', 'very bad', 'very good', 'good', 'bad', 'bad'], # Ordinality
        'shipping_provider': [f'Provider_{i}' for i in range(10)], # High cardinality
        'has_promo': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes'] # Target Variable
    }
    df = pd.DataFrame(data)
```
